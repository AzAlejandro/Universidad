## Pregunta 2

```python
#Se usará la función np.int64 de numpy aplicadas a los datos con apply()
data["cantidad"] = data["cantidad"].apply(np.int64)
data.dtypes
```

```python
#Además se creará la columna precio
data["precio"] = data["venta"]/data["cantidad"]
```

## Pregunta 3

```python
#Se realiza una copia de la data, se crea una columna de fechas para realizar los groupby
data3 = data.copy()
data3['fecha']=data[['dia','mes','año']].apply(lambda x : '{}-{}-{}'.format(x[0],x[1],x[2]), axis=1)
data3['fecha']=pd.to_datetime(data3['fecha'], format = "%d-%m-%Y")
#Se seleccionan las columnas a utilizar
data3 = data3[["fecha","cantidad","precio"]]
data3.head()
```

```python
data3["fecha2"] = data3["fecha"]  + pd.DateOffset(days = 1)
data3.sort_values("fecha", inplace = True)
```

```python
data3["esigual"] = (data3["fecha"].dt.month == data3["fecha2"].dt.month)
data3["esigual"] = np.where(data3["esigual"] == False, 1,0)
data3.tail(60)
```

```python
#Para crear el dataframe por mes, se agrupa por año y fecha y se realiza un aggregate según se pidió
df_month =data3.groupby([data3.fecha.dt.year,data3.fecha.dt.month]).agg({"fecha" : min,"cantidad":sum, "precio":"mean"})
df_month.reset_index(drop = True, inplace = True)
df_month.head()
```

```python
#Se realiza lo mismo, pero para indicar las semanas se usa .dt.isocalendar().week
df_week = data3.groupby([data3.fecha.dt.year,data3.fecha.dt.isocalendar().week]).agg({"fecha":min,"cantidad":sum, "precio":"mean"})
df_week.reset_index(drop = True, inplace = True)
df_week.tail()
```

```python
#Los datos ya venían por día, por lo cual solo se ordena por fecha y se arreglan los índices
df_day = data3.copy()
df_day.sort_values("fecha", inplace = True)
df_day.reset_index(drop = True, inplace = True)
df_day.head()
```

## Pregunta 4

## Pregunta 5

```python
#Se le da el formato a las fechas   "dd-mm-yyyy"
data5 = data4.copy()
data5_date = data5[["fecha","cantidad","precio"]]
data5_date = data5_date.sort_values("fecha")
data5_date = data5_date.reset_index(drop = True)
data5_date["fecha"] = data5_date.fecha.dt.strftime('%d-%m-%Y')
data5_date.tail(6)
```

## Pregunta 6

```python
#Se crea la columna "producto" con los datos del mani 
data6 = data5_date.copy()
sal = ["mani con sal 200 gr","mani sin sal 200 gr"]
data6['producto']= np.tile(sal, len(data6) // len(sal)).tolist() + sal[:len(data6)%len(sal)]
data6.reset_index(drop = True, inplace = True)
data6
```

## Pregunta 7

```python
#Maestra se carga localmente y se cambia el nombre descripcion a producto para luego realizar el merge
maestra = pd.read_excel("../Tarea1/maestra.xlsx")
maestra = maestra.rename(columns = {"descripcion" : "producto"})
maestra 
```

```python
#Se realiza un merge de las dos tablas
data7 = pd.merge(data6, maestra[["producto","sku"]])
#Se reemplazan los valores de sku del tipo XXXX-XX por solo la parte XXXX
data7["codigo"]=data7["sku"].map(lambda x: x.split("-")[0])
data7 = data7.drop("sku",1)
#Se renombra la columna descripcion que quedó del merge por producto
data7
```

## Pregunta 8

```python
#Al darle formato de dd-mm-yyyy está el problema que se convierte en string
#Al ser str sigue un orden distinto, por lo cual hay que hacer las conversiones
data8 = data7.copy()
data8['fecha']=pd.to_datetime(data8['fecha'], format ='%d-%m-%Y')
#Se ordenan los datos y se vuelve al formato original dd-mm-yyyy
data8.sort_values(["producto","fecha"], inplace = True)
data8.fecha = data8.fecha.dt.strftime('%d-%m-%Y')
data8 = data8.reset_index(drop = True)
data8
```

## Pregunta 9

```python
#Primero se copia la misma información de la pregunta 8 y se generan dos dataframes:
#data9_sin contiene los datos del producto sin sal
#data9_con contiene los datos del producto con sal
data9 = data8.copy()
data9['fecha']=pd.to_datetime(data9['fecha'], format = "%d-%m-%Y")
data9.sort_values("fecha", inplace = True)
data9_sin = data9[data9.producto == "mani sin sal 200 gr"]
data9_con = data9[data9.producto == "mani con sal 200 gr"]
data9_sin = data9_sin.reset_index(drop = True)
data9_con = data9_con.reset_index(drop = True)
```

```python
data9_con.head()
```

```python
data9_con.tail()
```

```python
#Se crea un bucle que compara la fecha con la fecha siguiente, si el mes cambia agrega un "1", 
#si se mantiene igual se agrega un "0". Finalmente como el último dato siempre será un cambio de fecha
#se hace un append.("1") a la lista. Esto se hace para el dataframe de los dos productos.
lista = []
for i in range(0, len(data9_con.fecha.dt.date)-1):    
    if data9_con.fecha.dt.date[i].month ==data9_con.fecha.dt.date[i+1].month:
        lista.append("0")
    else:
        lista.append("1")
lista.append("1")
data9_con.insert(5,"ultima fecha" ,lista)
```

```python
lista = []
for i in range(0, len(data9_sin.fecha.dt.date)-1):    
    if data9_sin.fecha.dt.date[i].month ==data9_sin.fecha.dt.date[i+1].month:
        lista.append("0")
    else:
        lista.append("1")
lista.append("1")
data9_sin.insert(5,"ultima fecha" ,lista)
```

```python
#Teniendo los dos dataframe con sus respectivos "ultima fecha" se hace un append para tener la union de los dataframe
data9_merg = data9_sin.append(data9_con)
data9_merg.sort_values(["producto","fecha"], inplace = True)
```

## Pregunta 10

```python
#Primero se crea una copia del dataframe de la pregunta 9
data10 = data9_merg.copy()
```

```python
#Se seleccionan los productos sin sal
data10_sin = data10[data10.producto == "mani sin sal 200 gr"].reset_index(drop = True)
#Se inicia una lista con valores None para agregar según se solicitó
lista = pd.Series([None])
#Se realiza un bucle for para localizar donde hay valores "ultima fecha" == 1. Esto se realiza un índice hacia atrás
#Para que el agregar una nueva fila no me diera problemas. 
#Luego se utiliza pd.concat(["datos anteriores","dato agregado","datos posteriores"]) para ir agregando las filas
#con valor None.
for i in range(1,len(data10_sin)):
    if data10_sin.loc[abs(i-1), "ultima fecha"] == "1":
        data10_sin = pd.concat([data10_sin.loc[:i-1], lista, data10_sin.loc[i:]])

data10_sin.loc[len(data10_sin)+1] = lista
    
#Por alguna razón se agregaba una fila que no tenía utilidad. Se filtran las columnas necesarias
data10_sin= data10_sin[["fecha","producto","cantidad","precio","ultima fecha"]]   
data10_sin =data10_sin.reset_index(drop = True)
```

```python
data10_sin.tail(60)
```

```python
#Para obtener la fecha donde realizar la media del precio se realiza un bucle for tomando el dato anterior a la fila con None
#Como el bucle se va realizando en un dataframe ordenado por fecha los meses y años se seleccionan
#correctamente. El bucle va ejerciendo acción cuando encuentra un valor nulo (pd.isna(data10_sin.loc[i, "producto"])).


for i in range(1,len(data10_sin)):
    if pd.isna(data10_sin.loc[i, "producto"]) :
        month = data10_sin.iloc[i-1,0].month
        year = data10_sin.iloc[i-1,0].year
        data10_sin.loc[i, "precio"] = data10_sin[(data10_sin.fecha.dt.month == month) & (data10_sin.fecha.dt.year == year)].precio.mean()


#Luego se transforma los datos de fecha a string para que agregar un valor 0 no diese problemas.
#Además se agrega el valor 2 para última fecha.        

data10_sin['fecha']=pd.to_datetime(data10_sin['fecha']).dt.strftime('%y-%m-%d')

for i in range(1,len(data10_sin)):
    if pd.isna(data10_sin.loc[i, "producto"]) :
        data10_sin.loc[i, "fecha"] = 0
        data10_sin.loc[i, "ultima fecha"] = 2

```
